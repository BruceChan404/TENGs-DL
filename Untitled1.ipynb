```python
def vectorgenerator(data,window):
    data=np.array(data)
#     maxval=data.max(axis=0)
#     maxtiming=data.argmax(axis=0)[maxval.argmax()]
    maxtiming=data.argmax(axis=0)[3]
    print(maxtiming)
    cut=data[maxtiming-window:maxtiming+100-window,:]
    cut=cut.T.reshape(1,900)
    return cut
def dataloader():
    trainX=np.zeros((100,900))
    trainY=np.zeros((100,1))
    print(trainX.shape)
    testX=np.zeros((25,900))
    testY=np.zeros((25,1))
    for i in range(0,5):
        for j in range(0,20):
            print(i,j)
            trainX[i*20+j,:]=vectorgenerator(dataset[classes[i]+str(j+1)],75)
            trainY[i*20+j]=i
        for j in range(20,25):
            print(i,j)
            testX[i*5+j-20,:]=vectorgenerator(dataset[classes[i]+str(j+1)],75)
            testY[i*5+j-20]=i
    return trainX,testX,trainY,testY
```

```python
trainX,testX,trainY,testY=dataloader()
```

```python
class ConvNet(torch.nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        
        ######  ########################################################################
        # TODO: Define a simple CNN contraining Conv, Pooling, and FC layers.        #
        ##############################################################################
        

        # Block 1:         1 x 30 x 30 --> 32 x 15 x 15        
        self.conv1 = torch.nn.Conv2d(1, 32, 3,padding=1) 
        # Block 2:         32 x 15 x 15 --> 64 x 16 x 16
        self.conv2 = torch.nn.Conv2d(32, 64, 2,padding=1)
        # Block 3:         64 x 16 x 16 --> 128 x 8 x 8        
        self.conv3 = torch.nn.Conv2d(64, 128, 3,padding=1)
        # Block 4:          128 x 8 x 8 --> 256 x 4 x 4
        self.conv4 = torch.nn.Conv2d(128, 256, 3,padding=1)
        # Linear layers:   256 x 4 x 4 --> 4096 --> 4096 --> 4096 --> 5
        self.fc1   = torch.nn.Linear(256*4*4, 4096) 
        self.fc2   = torch.nn.Linear(4096, 4096) 
        self.fc3   = torch.nn.Linear(4096, 4096)
        self.fc4   = torch.nn.Linear(4096, 5)
        
        ##############################################################################
        #                             END OF YOUR CODE                               #
        ##############################################################################

    def forward(self, x):
        
        ##############################################################################
        # TODO: Implement forward path turning an input image to class probability.  #
        # For activation function, please use ReLU.                                  #
        ##############################################################################

        # Block 1:         1 x 30 x 30 --> 16 x 15 x 15
        x = F.max_pool2d(F.relu(self.conv1(x)), (2, 2))
        # Block 2:         16 x 15 x 15 --> 32 x 8 x 8
        x = F.relu(self.conv2(x))
        # Block 3:         32 x 8 x 8 --> 64 x 4 x 4
        x = F.max_pool2d(F.relu(self.conv3(x)), (2, 2))
        # Block 4:         64 x 4 x 4 --> 128 x 2 x 2
        x = F.max_pool2d(F.relu(self.conv4(x)), 2)
        # Linear layers:   128 x 2 x 2 --> 512 --> 1024 --> 1024 --> 10
        x = x.view(x.size()[0], -1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = self.fc4(x)
        
        ##############################################################################
        #                             END OF YOUR CODE                               #
        ##############################################################################
        
        return x

model = ConvNet()
model.to(device)
print(model)
```

```python
def train_net(model):
    start=time.time()
    model.train()

    for epoch in range(epochs): 
        # epoch 0-5
        # set the running quatities to zero at the beginning of the epoch
        running_loss = 0
        running_error = 0
        for batch in range(5):
            inputs=train_X[20*batch:20*batch+20,:]
            labels=train_Y[20*batch:20*batch+20]
    #         print("inputs(a batch):",inputs.shape)
            pass
                ##############################################################################
                # TODO: Implement the training process.                                      #
                ##############################################################################

                # Load inputs and labels and deploy to running device
            inputs = inputs.to(device)
            labels = labels.to(device)
#             labels = labels.view(20,1)  # make it the same shape as output
                # Set the gradients to zeros
            optimizer.zero_grad()
                # Forward the batch data through the net       
            outputs = model(inputs)
    #         print("outputs:",outputs.shape)
    #         print("labels:",labels.shape)
    #         print(outputs)
    #         print(labels)
                # Compute the average of the losses of the data points in the minibatch
            loss = criterion(outputs, labels)
                # Backward pass to compute gradients
            loss.backward()
                # Do one step of stochastic gradient descent
            optimizer.step()
                # Add the loss of this batch to the running loss
            running_loss += loss.item()
                # Compute the error made on this batch and add it to the running error
                    #find the max score for 10labels and make prediction
            p_labels=torch.max(outputs,1)[1]
#             print(p_labels)
# #             print(labels)
#             print(p_labels==labels.T)
#             print((p_labels==labels.T).sum())
#             print(20-(p_labels==labels.T).sum())
                    #check every prediction are wrong/right and calculate the wrong number in this batch 
            wrong_num=20-(p_labels==labels).sum() 
#             print("w_num=",wrong_num)
                    #add to running_error
            running_error += wrong_num
            ##############################################################################
            #                             END OF YOUR CODE                               #
            ##############################################################################
            
        # Compute stats for the full training set
        total_loss = running_loss / train_size
        total_error = running_error / train_size
        elapsed = (time.time()-start) / 60
        
        print('epoch= {} \t time= {:.2f} min \t loss= {:.3f} \t error= {:.2f}%'.format(epoch, elapsed, total_loss, total_error * 100))
        eval_on_test_set(model,epoch)
```
